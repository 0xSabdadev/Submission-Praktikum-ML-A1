```python
# Mengimport modul pandas
import pandas as pd
```

1. Data yang digunakan

```python
# Variabel names berisi daftar nama kolom/feature yang akan digunakan untuk DataFrame yang akan dibaca.
names = ['RI', 'Na', 'Mg', 'AI', 'Si', 'K', 'Ca', 'Ba', 'Fe', 'Type of glass']

# Keterangan setiap feature

# 1. RI (Refractive Index)
# Ini adalah indeks bias suatu material terhadap cahaya. Ini adalah ukuran sejauh mana cahaya dibiaskan saat melewati suatu material.
# Nilai RI untuk berbagai jenis kaca dapat bervariasi, dan ini adalah salah satu fitur yang digunakan untuk membedakan jenis kaca.

# 2. Na (Sodium)
# Kandungan natrium dalam kaca. Kandungan natrium dapat bervariasi antara jenis kaca, dan itu bisa menjadi fitur penting dalam identifikasi.

# 3. Mg (Magnesium)
# Kandungan magnesium dalam kaca. Seperti Na, kandungan magnesium juga dapat bervariasi antara jenis kaca.

# 4. AI (Aluminum)
# Kandungan aluminium dalam kaca. Kandungan aluminium dapat memberikan petunjuk tentang jenis kaca yang sedang diidentifikasi.

# 5. Si (Silicon)
# Kandungan silikon dalam kaca. Silikon adalah komponen umum dalam kaca, tetapi kandungannya dapat bervariasi.

# 6. K (Potassium)
# Kandungan potassium dalam kaca. Seperti fitur lainnya, kandungan potassium dapat berbeda antara jenis kaca.

# 7. Ca (Calcium)
# Kandungan kalsium dalam kaca. Kalsium adalah komponen umum dalam banyak jenis kaca.

# 8. Ba (Barium)
# Kandungan barium dalam kaca. Barium adalah elemen yang dapat digunakan sebagai fitur pemisah dalam identifikasi jenis kaca.

# 9. Fe (Iron)
# Kandungan besi dalam kaca. Besi juga dapat menjadi fitur yang penting dalam identifikasi, karena kandungannya bervariasi antara jenis kaca.

# 10. Type of glass
# Type of glass ini bukan merupakan sebuah fitur pada dataset tersebut, namun merupakan hasil prediksi dari jenis kaca.
# Kolom ini dimasukkan karena apabila hanya 9 fitur saja yang ditampilkan akan merubah hasil dari data yang ada.
# Jenis-jenis kaca :
#     -- 1 building_windows_float_processed
#     -- 2 building_windows_non_float_processed
#     -- 3 vehicle_windows_float_processed
#     -- 4 vehicle_windows_non_float_processed (none in this database)
#     -- 5 containers
#     -- 6 tableware
#     -- 7 headlamps

# Dataset yang digunakan
dataset = pd.read_csv('glass.data', names=names)
```

Berikut merupakan bagian-bagian dari box and whisker plotâ–¶:

- Minimum: Nilai minimum dalam kumpulan data yang diberikan. Minimum adalah Q1 - 1.5 * IQR dan ditunjukkan oleh garis (whisker) sebelah bawah dari median (setelah kurtil ketiga).

- Kuartil Pertama (Q1): Kuartil pertama adalah median setengah bawah dari kumpulan data.

- Median: Median adalah nilai tengah dari kumpulan data, yang membagi kumpulan data yang diberikan menjadi dua bagian yang sama. Median dianggap sebagai kuartil kedua. Median ditunjukkan oleh garis yang berada di tengah kotak(box).

- Kuartil Ketiga (Q3): Kuartil ketiga adalah median setengah atas dari data.

- Maksimum: Nilai maksimum dalam kumpulan data yang diberikan. Maksimum adalah Q3 - 1.5 * IQR dan ditunjukkan oleh garis (whisker) sebelah atas dari median (sebelum kuartil pertama).

- Rentang Antarkuartil (IQR): Perbedaan antara kuartil ketiga dan kuartil pertama dikenal sebagai rentang antarkuartil, yaitu IQR = Q3-Q1. IQR ditunjukkan oleh kotak(box).

- Outlier: Data yang jatuh di sisi kiri atau kanan yang jauh dari data yang diurutkan diuji untuk menjadi outlier. Umumnya, outlier jatuh lebih dari jarak yang ditentukan dari kuartil pertama dan kuartil ketiga. Outlier ditunjukkan oleh lingkaran kecil yang berada diuar box and whisker. Outlier lebih besar dari Q3 + (1,5 . IQR) atau kurang dari Q1 - (1,5 . IQR).
