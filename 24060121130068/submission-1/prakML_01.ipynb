### e. visualisasi data menggunakan plot unvariant

```python
# visualisasi data menggunakan plot unvariant (box diagram)
# mengimport matplotlib.pyplot
import matplotlib.pyplot as plt

# Ubah ukuran figur
fig = plt.figure(figsize=(12, 10))

# Membuat subplot
# kind: jenis diagram
# subplots=True: membuat subplot terpisah untuk setiap fitur dalam dataset.
# layout: mengatur tata letak subplot (baris, kolom).
# sharex=False dan sharey=False: antar subplot tidak akan memiliki sumbu x dan y yang sama.
# ax=fig.gca() mengacu pada objek "fig" yang telah dibuat sebelumnya.
dataset.plot(kind='box', subplots=True, layout=(4, 4), sharex=False, sharey=False, ax=fig.gca())

# Mengatur jarak antara subplot
plt.tight_layout()

# Menampilkan diagram box
plt.show()
```

### Penjelasan:
Plot univariat adalah plot dari masing-masing variabel individu.
Mengingat bahwa variabel inputnya numerik, kita bisa membuat jenis
plot box. Interpretasi dari box plot bergantung pada tujuan analisis. Box plot membantu dalam memahami sebaran data, nilai median, serta keberadaan pencilan. Ini adalah alat yang berguna dalam analisis eksploratori data dan membandingkan sekelompok data.

Berikut yang dapat diinterpretasi dari diagram box:
Interpretasi dari sebuah box plot melibatkan pemahaman terhadap berbagai elemen dalam plot tersebut. Berikut adalah cara untuk menginterpretasi box plot:

- Kotak (Box): Kotak pada box plot mewakili interkuartil range (IQR), yang mencakup sekitar 50% data di tengah distribusi. Batas bawah kotak menunjukkan nilai kuartil pertama (Q1), sedangkan batas atas kotak menunjukkan nilai kuartil ketiga (Q3). Kotak itu sendiri melambangkan sebaran data antara Q1 dan Q3.

- Garis Tengah di Kotak: Garis tengah di dalam kotak adalah nilai median (Q2) dari data, yang membagi data menjadi dua bagian yang sama dari segi jumlah.

- Whisker Atas dan Whisker Bawah: Garis-garis yang menghubungkan kotak ke titik-titik di atas dan di bawahnya menunjukkan jangkauan data di luar IQR. Garis-garis ini biasanya ditarik hingga batas atas dan batas bawah data yang tidak dianggap sebagai pencilan. Pencilan mungkin muncul sebagai titik-titik di luar garis whiskers.

- Pencilan (Outliers): Titik-titik di luar batas atas dan batas bawah garis whiskers adalah pencilan. Pencilan adalah data-data yang jauh dari nilai-nilai utama distribusi dan perlu diperhatikan dalam analisis data.

- Posisi dan Panjang Kotak: Posisi kotak di sepanjang sumbu horizontal menunjukkan letak pusat distribusi data, sedangkan panjang kotak menggambarkan sebaran data dalam IQR.

- Arah dan Panjang whiskers: Arah dan panjang garis whisker menunjukkan sejauh mana data menyebar di luar IQR. Jika whisker lebih panjang, ini menunjukkan distribusi data yang lebih luas.

- Skewness adalah konsep dalam statistik yang mengacu pada kemiringan atau asimetri dalam distribusi data. Ini mengukur sejauh mana distribusi data cenderung condong ke satu arah, baik ke kanan (positif skewness) atau ke kiri (negatif skewness), atau apakah distribusi data cenderung simetris. Terdapat tiga jenis skewness sebagai berikut

 a. Positif (Right-Skewed): Jika data cenderung miring ke kanan, maka distribusinya dikatakan positif atau right-skewed. Ini berarti sebagian besar data terkumpul di sebelah kiri dan upper whisker cenderung lebih panjang dari lower whisker karena ekor yang lebih panjang di sebelah kanan.

 b. Negatif (Left-Skewed): Jika data cenderung miring ke kiri, maka distribusinya dikatakan negatif atau left-skewed. Ini berarti sebagian besar data terkumpul di sebelah kanan dan lower whisker cenderung lebih panjang dari upper whisker karena ekor yang lebih panjang di sebelah kiri.

 c. Simetris (No Skew): Jika data terdistribusi secara simetris, maka distribusinya dikatakan tidak miring (no skew). Ini berarti data terdistribusi secara merata di sekitar nilai tengahnya, dan skewnessnya mendekati nol.

### f. visualisasi data menggunakan plot multivarian

```python
# visualisasi data menggunakan plot multivarian (matriks scatter)
# import scatter_matrix
from pandas.plotting import scatter_matrix

# Membuat figur dengan ukuran lebih besar
fig = plt.figure(figsize=(16, 14))

# Membuat matriks scatter dengan subplot
# alpha: mengatur tingkat transparansi titik-titik pada scatter plot.
# figsize: mengatur ukuran gambar matriks scatter (baris, kolom)
# diagonal='hist': mengatur bahwa elemen diagonal dalam matriks akan berisi histogram untuk setiap fitur.
# color:mengatur warna plot scatter.
scatter_matrix(dataset, alpha=0.5, figsize=(16, 14), diagonal='hist', color='skyblue')

# Menampilkan matriks scatter
plt.show()
```

### Penjelasan:
kita bisa melihat interaksi antar variabel. Pertama, kita lihat scatterplots dari semua pasang atribut. Hal ini dapat membantu melihat hubungan terstruktur antara variabel input (pola hubungan), arah hubungan, dan keeratan hubungan.

- Pola Hubungan: Scatterplot memungkinkan kita untuk mengidentifikasi pola hubungan antara dua variabel. Pola dapat berupa linear (garis lurus), eksponensial (meningkat secara eksponensial), logaritmik (menurun secara logaritmik), atau pola non-linear lainnya. Dengan memperhatikan pola ini, kita dapat mengidentifikasi bagaimana satu variabel berubah seiring dengan perubahan variabel lainnya.

- Arah Hubungan: Scatterplot juga membantu kita menentukan arah hubungan antara dua variabel. Jika titik-titik cenderung bergerak dari kiri ke kanan saat salah satu variabel meningkat, ini menunjukkan hubungan positif, yang berarti bahwa kenaikan dalam satu variabel cenderung disertai dengan kenaikan dalam variabel lainnya. Sebaliknya, jika titik-titik cenderung bergerak dari kiri atas ke kanan bawah saat salah satu variabel meningkat, ini menunjukkan hubungan negatif, yang berarti bahwa kenaikan dalam satu variabel disertai dengan penurunan dalam variabel lainnya.

- Keeratan Hubungan: Keeratan hubungan antara dua variabel dapat diamati dengan melihat seberapa erat titik-titik berdekatan satu sama lain dalam scatterplot. Semakin erat titik-titik berdekatan, semakin erat hubungan antara variabelnya. Jika titik-titik tersebar dengan lebar dan tidak mengikuti pola tertentu, ini menunjukkan hubungan yang lebih lemah.

- Pencilan (Outliers): Scatterplot juga membantu mengidentifikasi pencilan (outliers) dalam data. Pencilan adalah titik-titik yang terletak jauh dari pola hubungan umum antara dua variabel. Identifikasi pencilan dapat memberikan wawasan tentang anomali dalam data.
