Selanjutnya akan melakukan penentuan dataset denagn url yang diagunakan, dengan melakukan pendefinisian

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/zoo/zoo.data"
```

kemudian menetukan fitur yang akan digunakan pada dataset kita, dataset zoo memeiliki 18 fitur.

```python
names=['animal_name','hair','feathers','eggs','milk','airborne','aquatic','predator','toothed','backbone','breathes','venomous','fins','legs','tail','domestic','catsize','type']
```

lalu ini akan membaca dataset tersebut  menjadi sebuah tabel dan akan dilihat dimensinya

```python
dataset = pd.read_csv(url, names=names)
```

```python
print(dataset.shape)
```
