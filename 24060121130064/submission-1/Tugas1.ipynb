```python
# import library pandas yang digunakan untuk mengelola dataset
import pandas as pd
```

```python
# Memuat data set

# url disini merupakan parameter untuk menyimpan link yang merujuk datasetnya
url = "https://archive.ics.uci.edu/static/public/9/data.csv"

# names disini disini merupakan parameter untuk menyimpan array nama kolomnya nanti
names = ['car_name', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'mpg']

# param dataset disini digunakan untuk menyimpan data dari dataset yang sudah ada di param url yang dibaca dengan fungsi .read_csv yang dimiliki library pandas
dataset = pd.read_csv(url)

#mengubah nama kolom sesuai parameter names
dataset.rename(columns={'car_name': names[0], 'cylinders': names[1], 'displacement': names[2], 'horsepower': names[3], 'weight': names[4], 'acceleration': names[5], 'model_year': names[6], 'origin': names[7], 'mpg': names[8]}, inplace=True)
```

#Melakukan Summary Dataset

```python
# I. Menentukan Dimensi dari Dataset

# kode dibawah ini akan menampilkan dimensi dari dataset tersebut dimana memberikan nilai banyak baris dan banyak kolom

dataset.shape
# banyak barisnya 398
# banyak kolom 9
```

```python
# II. Melihat Isi Dataset

# kode dibawah akan menampilkan 20 baris pertama dari dataset
dataset.head(20)
```

```python
# III. Distribusi Kelas Data

# kode dibawah akan menampilkan banyak nya nilai untuk setiap value yang ada di kolom cylinders
dataset.groupby('cylinders').size()
```

```python
# IV. Ringkasan Statistik

#kode Berikut adalah perintah untuk mendeskripsikan statistik data meliputi count, mean, standard deviasi, nilai min, nilai max, dan quartile
dataset.describe()
```

```python
# V. Visualisasi Data Menggunakan Plot Univariat
# a. plot box

# import matplotlib dengan memberinya alias 'plt' untuk membuat visualisasi data, termasuk berbagai jenis plot dan bagan
import matplotlib.pyplot as plt


dataset.plot(kind = 'box', subplots = True, layout=(20,3), sharex = False, sharey = False, figsize = (12, 80))
# pada kode diatas ini menggunakan fungsi .plot dengan beberapa parameter didalamnya antara lain:
# kind adalah parameter yang menentukan jenis plot yang digunakan, yang dalam hal ini adalah plot box. Plot box menampilkan distribusi data, menunjukkan median, kuartil, dan potensi outlier.
# subplots Jika disetel ke True, parameter ini membuat subplot terpisah untuk setiap kolom numerik di DataFrame, memungkinkan Anda memvisualisasikan beberapa plot kotak secara berdampingan.
# layout  Parameter ini menentukan tata letak subplot. Dalam hal ini, ini membuat petak subplot berukuran 20x3
# sharex Parameter ini mengontrol apakah sumbu x dibagikan di antara subplot. Jika disetel ke False, setiap subplot memiliki sumbunya sendiri-sendiri.
# sharey  Parameter ini mengontrol apakah sumbu y dibagi di antara subplot. Jika disetel ke False, setiap subplot memiliki sumbunya sendiri-sendiri.
# figsize Parameter ini menetapkan ukuran keseluruhan gambar (seluruh grid subplot) menjadi lebar 12 unit dan tinggi 80 unit.

plt.show()
# kode diatas digunakan untuk menampilkan bagan nya
```

```python
# b. histogram

# kode dibawah ini digunakan untuk membuat bagan dalam bentuk histogram untuk setiap kolom yang ada di dataset. Histogram merupakan graph yang menampilkan persebaran dataset
# menampilkan frekuensi dari nilai menggunakan batas interval
dataset.hist(figsize = (12, 10))

# digunakan untuk menampilkan bagan
plt.show()
```

```python
# VI. Visualisasi Data Menggunakan Plot Multivariat

# import pandas yang digunakan untuk memanipulasi data dan menganalisisnya
# lalu menggunakan meng import scatter_matrix yang digunakan untuk membuat scatter matrix plot
import pandas
from pandas.plotting import scatter_matrix

# parameter yang digunakan dalam fungsi scatter_matrix adalah :
# dataset merupakan parameter untuk menyimpan nilai dataset yang digunakn untuk data yang akan di tampilkan.
# figsizemerupakan parameter untuk mengatur ukuran untuk setiap plot scatter matrix yang ditampilkan, dimensi yang digunakan yaitu 15 X 15 inches
scatter_matrix(dataset, figsize=(15, 15))

# digunakan untuk menampilkan matrixnya
plt.show()
```
