```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
#import dataset dari sklearn
from sklearn.datasets import load_diabetes
```

```python
#memuat dataset diabetes
diabetes = load_diabetes()
diabetes_df = pd.DataFrame(data=diabetes.data, columns=diabetes.feature_names)
diabetes_df['class'] = diabetes.target
diabetes_df.head(10)
```

```python
# Separating out the features
x = diabetes_df.loc[:, diabetes.feature_names].values
# Separating out the target
y = diabetes_df.loc[:, ['class']].values
x = StandardScaler().fit_transform(x)
pd.DataFrame(data = x, columns = diabetes.feature_names).head()
```

x = diabetes_df.loc[:, diabetes.feature_names].values: Ini adalah langkah pertama dalam memisahkan fitur dari dataset. diabetes_df adalah DataFrame pandas yang berisi data diabetes. diabetes.feature_names digunakan untuk memilih kolom-kolom yang merupakan fitur-fitur dari dataset. loc[:, diabetes.feature_names] mengambil semua baris (:) dari kolom-kolom yang sesuai dengan nama fitur diabetes. .values kemudian mengubah DataFrame menjadi array NumPy, sehingga variabel x sekarang berisi matriks fitur.

y = diabetes_df.loc[:, ['class']].values: Ini adalah langkah untuk memisahkan target dari dataset. 'class' mungkin merupakan kolom yang berisi label kelas atau variabel target. loc[:, ['class']] digunakan untuk memilih kolom 'class'. Seperti sebelumnya, .values mengubah DataFrame menjadi array NumPy, sehingga variabel y berisi array target.

x = StandardScaler().fit_transform(x): Ini adalah langkah normalisasi atau standarisasi fitur. StandardScaler dari scikit-learn digunakan untuk mengubah distribusi nilai fitur agar memiliki rata-rata 0 dan deviasi standar 1. Ini membantu algoritma pembelajaran mesin yang sensitif terhadap skala untuk konvergensi lebih cepat dan memberikan bobot yang lebih baik. fit_transform digunakan untuk menghitung rata-rata dan deviasi standar dari data (fit) dan kemudian mengubah data (transform) berdasarkan nilai-nilai tersebut.

pd.DataFrame(data=x, columns=diabetes.feature_names).head(): Ini hanya digunakan untuk membuat DataFrame pandas baru dari data yang sudah dinormalisasi dan menampilkan beberapa baris pertama dari DataFrame tersebut menggunakan .head(). Ini membantu untuk memastikan bahwa normalisasi telah dilakukan dengan benar dan untuk melihat bagaimana data terlihat setelah proses normalisasi.

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf.head(5)
```


Kode di atas menggunakan Analisis Komponen Utama (PCA) untuk mengurangi dimensi dari dataset. Dengan menciptakan objek PCA dan mengatur jumlah komponen utama yang diinginkan menjadi dua, kode tersebut melakukan transformasi PCA pada matriks fitur. Hasilnya, disimpan dalam DataFrame baru yang menunjukkan dua komponen utama sebagai kolom 'principal component 1' dan 'principal component 2'. Proses PCA membantu mengurangi kompleksitas data, mempertahankan sebanyak mungkin informasi, dan mengubah data ke dalam ruang fitur yang lebih rendah dimensi. DataFrame hasilnya memberikan gambaran awal tentang representasi data setelah proses reduksi dimensi oleh PCA.

```python
finalDf = pd.concat([principalDf, diabetes_df['class']], axis = 1)
finalDf.head(10)
```

Kode di atas menggabungkan hasil transformasi PCA dengan kolom target dari dataset diabetes untuk membentuk DataFrame akhir.

finalDf = pd.concat([principalDf, diabetes_df['class']], axis=1): Menggunakan fungsi concat dari pandas untuk menggabungkan dua DataFrames, yaitu principalDf yang berisi dua komponen utama hasil transformasi PCA, dan kolom 'class' dari DataFrame asli diabetes_df. Parameter axis=1 menunjukkan bahwa penggabungan dilakukan secara horizontal (sepanjang kolom). Hasilnya adalah DataFrame baru yang disimpan dalam variabel finalDf.

finalDf.head(10): Menampilkan sepuluh baris pertama dari DataFrame finalDf. Ini memberikan gambaran awal tentang bagaimana representasi data terlihat setelah transformasi PCA dan penggabungan dengan kolom target. Setiap baris mencakup nilai dari dua komponen utama ('principal component 1' dan 'principal component 2') serta nilai dari kolom target ('class').

Hasil akhir, yaitu finalDf, menyatukan informasi dari hasil PCA dengan informasi target, yang dapat digunakan untuk analisis lebih lanjut atau pemodelan prediktif. Integrasi komponen utama dengan label kelas membantu melihat hubungan antara fitur-fitur utama dan target kelas dalam dimensi yang lebih rendah setelah proses PCA.

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = [118.0, 144.0, 97.0, 68.0]
colors = ['r', 'g', 'b', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['class'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

Kode pca.explained_variance_ratio_ digunakan untuk mendapatkan rasio varians yang dijelaskan oleh setiap komponen utama hasil PCA

pca.explained_variance_ratio_: Memanggil atribut explained_variance_ratio_ dari objek PCA (pca). Atribut ini memberikan nilai rasio varians yang dijelaskan oleh setiap komponen utama. Dalam hal ini, metode ini mengembalikan array NumPy yang berisi dua angka, yang masing-masing mewakili proporsi varians yang dijelaskan oleh komponen utama pertama dan kedua.

Hasilnya adalah array: [0.40242108, 0.14923197]. Penjelasannya sebagai berikut:

Komponen utama pertama (PC1) menjelaskan sekitar 40.24% dari varians total dalam data.
Komponen utama kedua (PC2) menjelaskan sekitar 14.92% dari varians total dalam data.
Rasio varians ini memberikan informasi tentang sejauh mana masing-masing komponen utama mewakili variasi dalam data. Dengan melihat hasil ini, kita dapat menilai seberapa baik dua komponen utama tersebut mampu menjelaskan variasi dalam data asli. Dalam hal ini, sekitar 55.16% dari total varians dapat dijelaskan oleh kedua komponen utama. Sisanya, sekitar 44.84% varians lainnya mungkin tersebar di komponen utama lainnya (jika ada) atau tetap ada dalam fitur-fitur yang tidak dimasukkan dalam analisis PCA ini.
