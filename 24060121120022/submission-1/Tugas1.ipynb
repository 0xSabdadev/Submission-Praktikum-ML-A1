```python
# Importing the necessary function and library:
# ucimlrepo: Loads the dataset from the UCI ML Repository, including the dataframes and metadata information
# pandas: A software library written for data manipulation and analysis as pd

from ucimlrepo import fetch_ucirepo
import pandas as pd

# Fetching the dataset National Health and Nutrition Health Survey 2013-2014 (NHANES) Age Prediction Subset
# This dataset has 10 attributes and needs to be renamed to make it readable with an array of names
url = "https://archive.ics.uci.edu/static/public/887/data.csv"
names = ['ID', 'age_group', 'age', 'gender', 'activity_level', 'bmi', 'blood_glucose', 'is_diabetic', 'ogtt', 'insulin_levels']

# Reading the csv file from the URL and generating a DataFrame via read_csv function
dataframe = pd.read_csv(url)

# Renaming the columns with the array of names
dataframe.rename(columns={'SEQN': names[0], 'age_group': names[1], 'RIDAGEYR': names[2], 'RIAGENDR': names[3], 'PAQ605': names[4], 'BMXBMI': names[5], 'LBXGLU': names[6], 'DIQ010': names[7], 'LBXGLT': names[8], 'LBXIN': names[9]}, inplace=True)
```

```python
# 5. Visualizing The Data Using The Univariate Plot
# a. With Box Plot

# Importing the matplotlib library and assigns it an alias 'plt' for creating data visualizations, including various types of plots and charts
import matplotlib.pyplot as plt

# Creating a box plots with these parameters:
# kind = 'box'      : This specifies the type of plot to create, which is a box plot in this case. A box plot displays the distribution of data, showing the median, quartiles, and potential outliers.
# subplots = True   : When set to True, this parameter creates separate subplots for each numerical column in the DataFrame, allowing you to visualize multiple box plots side by side.
# layout = (20,3)   : This parameter specifies the layout of the subplots. In this case, it creates a 20x3 grid of subplots, meaning four subplots in total.
# sharex = False    : These parameters control whether the x-axis are shared among the subplots. When set to False, each subplot has its own distinct axis.
# sharey = False    : These parameters control whether the y-axis are shared among the subplots. When set to False, each subplot has its own distinct axis.
# figsize = (10,80) : This parameter sets the size of the entire figure (the entire grid of subplots) to be 10 units in width and 80 units in height.

dataframe.plot(kind = 'box', subplots = True, layout=(20,3), sharex = False, sharey = False, figsize = (10, 80))

# Displaying the created box plots. Without this line, the plots would not be visible.
plt.show()
```

As shown in the DataFrame description in 4a (median, quartiles, min, max), here are the explanation of the output above, indicating potential outliers for each attribute:

```python
# b. With Hstogram

# Creating histograms for all the columns in the DataFrame with 'figsize = (12, 8)' to set the overall size of the figure.
# A histogram is a graphical representation of the distribution of a dataframe, showing the frequencies of values within specified intervals or bins
dataframe.hist(figsize = (12, 12))

# Displaying the generated histograms. Without this line, the plots would not be visible.
plt.show()
```

```python
# 6. Visualizing The Data Using The Multivariate Plot
# Using Scatter Matrix

# Importing the Pandas library to be used for data manipulation and analysis and also
# scatter_matrix function from the Pandas plotting module to be used to create scatter matrix plots.
import pandas
from pandas.plotting import scatter_matrix

# Creating the Scatter Matrix with these parameters:
# dataframe       : Passing the DataFrame as the data to be visualized.
# figsize=(10,10) : Controls the size of the entire scatter matrix plot means that the plot will have dimensions of 10x10 inches.
scatter_matrix(dataframe, figsize=(10, 10))
plt.show()
```

```python
# Using Seaborn Pairplot

# Importing the Seaborn library to provide a high-level interface for creating aesthetically pleasing and informative statistical graphics
import seaborn

# Creating the pairplot function from Seaborn, passing DataFrame as the data to be visualized.
# The pairplot function automatically creates a grid of scatter plots, where each cell in the grid represents a scatter plot of two numerical variables from the dataframe.
seaborn.pairplot(dataframe)
```
