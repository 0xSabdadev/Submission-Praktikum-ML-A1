```python
import pandas as pd
```

```python
#untuk menginisialisasi variable "url" dengan URL dari dataset Abalone
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"

#menginisialisasi variable 'names' dengan daftar nama kolom yang akan digunakan untuk dataset Abalone
names = ['Sex', 'Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings']

#pustaka pandas untuk membaca dataset dari URL yang telah ditentukan sebelumnya
dataset = pd.read_csv(url, names=names)
```

```python
#untuk menentukan dimensi dari dataset
print(dataset.shape)
```

```python
#Melihat isi dataset keseluruhan
print(dataset)
```

```python
#Melihat isi dataset 20 paling akhir
print(dataset.tail(20))
```

```python
#Distribusi Kelas Data
dataset.groupby('Rings').size()
```

```python
#Ringkasan Statistik
dataset.describe()
```

```python
#Visualisasi Data Menggunakan Plot Univariat
import matplotlib.pyplot as plt
dataset.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

```python
#Membuat Histogram masing-masing variabel input
dataset.hist()
plt.show()
```

```python
#Visualisasi Data menggunakan plot multivariat
import pandas
from pandas.plotting import scatter_matrix
scatter_matrix(dataset)
plt.show()
```
