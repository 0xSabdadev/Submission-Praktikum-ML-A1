```python
# Impor library yang dibutuhkan
import pandas as pd
```

```python
# Memuat dataset dari url yang telah dipilih
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/balance-scale/balance-scale.data"

names = ['class name', 'left-weight', 'left-distance', 'right-weight', 'right-distance']

dataset = pd.read_csv(url, names=names)
```

```python
# Mengganti nilai 'L', 'B', 'R' dengan 1, 2, 3 dalam kolom 'class name'
class_mapping = {'L': 1, 'B': 2, 'R': 3}
dataset['class name'] = dataset['class name'].map(class_mapping)
```

```python
# Melihat dimensi dataset
dataset.shape
```

```python
# Melihat isi dataset
dataset
```

```python
# Mendistribusikan kelas data
dataset.groupby('class name').size()
```

```python
# Menampilkan ringkasan statistik dataset
dataset.describe()
```

```python
# Menampilkan informasi dari dataset
dataset.info()
```

```python
# Import library yang dibutuhkan
import matplotlib.pyplot as plt
```

```python
# Membuat box plot dari dataset
dataset.plot(kind='box', subplots=True, layout=(2,3), sharex=False, sharey=False)
plt.show()
```

Plot tersebut menggunakan boxplot untuk menunjukkan distribusi data dari masing-masing variabel.

Berdasarkan plot tersebut :

1. Terdapat 3 kelas, yaitu kelas 1.0, kelas 2.0, dan kelas 3.0.

2. Kelas 3.0 merupakan berat ke kanan, kelas 2.0 merupakan titik seimbang, dan kelas 1.0 merupakan berat ke kiri.

Secara keseluruhan, plot univariat tersebut memberikan gambaran umum mengenai karakteristik dari data yang dianalisis. Plot tersebut dapat digunakan untuk mengidentifikasi pola dan tren dalam data, serta untuk membandingkan karakteristik antar kelompok data.



```python
# Menampilkan data dalam bentuk histogram
dataset.hist(layout=(2,3))
plt.show()
```

Dari histogram tersebut dapat dilihat pada class name jumlah data yang berat ke kanan dan ke kiri memiliki frekuensi yang sama sedangkan pada data yang seimbang hanya memiliki frekuensi yang kurang dari 100.
