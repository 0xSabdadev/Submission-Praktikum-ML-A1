```python
# Membaca file CSV untuk kedua dataset dan menetapkan nama kolom
names = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol", "quality"]
dataset_red = pd.read_csv('/content/drive/MyDrive/ML/MLA1/winequality-red.csv', delimiter=';', names=names,header=0)
dataset_white = pd.read_csv('/content/drive/MyDrive/ML/MLA1/winequality-white.csv', delimiter=';', names=names,header=0)
```

Kode di atas membaca dua file CSV yang berisi data kualitas anggur, memuatnya ke dalam dua dataset terpisah: dataset_red untuk anggur merah dan dataset_white untuk anggur putih.

Pada baris pertama, digunakan untuk daftar nama kolom yang akan digunakan untuk dataset. Nama-nama ini akan digunakan sebagai header atau label kolom untuk setiap dataset yang akan dibaca.



```python
dataset_red['type'] = 'red'
dataset_white['type'] = 'white'
```

Pada kode di atas, kita menambahkan kolom baru bernama 'type' ke dalam kedua dataset yang telah dibaca sebelumnya: dataset_red dan dataset_white. Kita menetapkan nilai 'red' pada kolom 'type' di dataset_red dan nilai 'white' pada kolom 'type' di dataset_white.

```python
# Menggabungkan kedua dataset
combined_dataset = pd.concat([dataset_red, dataset_white])
```

Pada kode di atas, kita menggabungkan kedua dataset, dataset_red (yang berisi data tentang anggur merah) dan dataset_white (yang berisi data tentang anggur putih) menjadi satu dataset tunggal yang disebut combined_dataset.

```python
import matplotlib.pyplot as plt

# Menentukan ukuran gambar keseluruhan
fig = plt.figure(figsize=(40, 40))

# Menentukan ukuran setiap subplot
subplot_size = (2, 2)

# Melakukan plotting dengan subplot dan mengatur ukuran subplot
for i in range(1, 5):
    ax = fig.add_subplot(2, 2, i)
    combined_dataset.plot(kind='box', ax=ax, subplots=True, layout=(10, 10), sharex=False, sharey=False)

# Menampilkan plot
plt.show()
```

```python
fig, axs = plt.subplots(2, 7, figsize=(20, 10))  # Membuat subplot 10x10

# Flatten array dari subplot untuk memudahkan iterasi
axs = axs.flatten()

# Loop untuk membuat histogram untuk setiap kolom di dataset
for i, column in enumerate(combined_dataset.columns):
    combined_dataset[column].hist(ax=axs[i])  # Menempatkan histogram ke subplot yang sesuai
    axs[i].set_title(column)  # Menambahkan judul untuk setiap subplot

plt.tight_layout()  # Mengatur tata letak agar tidak tumpang tindih
plt.show()
```

Membuat Subplot dan Pengaturan Awal:

fig, axs = plt.subplots(2, 7, figsize=(20, 10)) berusaha membuat subplot dengan ukuran 2 baris dan 7 kolom serta ukuran total 20x10 inch.
axs = axs.flatten() mengubah array dari subplot ke dalam satu dimensi untuk memudahkan iterasi.
