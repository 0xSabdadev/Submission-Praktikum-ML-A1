# Import Libraries

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
```

# Memuat Dataset

```python
# Memuat Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/nursery/nursery.data"

names = ['parents', 'has_nurs', 'form', 'children', 'housing', 'finance', 'social', 'health', 'class']

dataset = pd.read_csv(url, names=names)
```

# Mengubah Data Type dari Categorical menjadi Numerical

```python
# Mengubah Data Type dari Categorical menjadi Numerical menggunakan fungsi map()
# agar dapat dibuat bentuk Plotnya

# Mapping for 'parents'
parents_mapping = {'usual': 1, 'pretentious': 2, 'great_pret': 3}
dataset['parents'] = dataset['parents'].map(parents_mapping)

# Mapping for 'has_nurs'
has_nurs_mapping = {'proper': 1, 'less_proper': 2, 'improper': 3, 'critical': 4, 'very_crit': 5}
dataset['has_nurs'] = dataset['has_nurs'].map(has_nurs_mapping)

# Mapping for 'form'
form_mapping = {'complete': 1, 'completed': 2, 'incomplete': 3, 'foster': 4}
dataset['form'] = dataset['form'].map(form_mapping)

# Mapping for 'children'
children_mapping = {'1': 1, '2': 2, '3': 3, 'more': 4}
dataset['children'] = dataset['children'].map(children_mapping)

# Mapping for 'housing'
housing_mapping = {'convenient': 1, 'less_conv': 2, 'critical': 3}
dataset['housing'] = dataset['housing'].map(housing_mapping)

# Mapping for 'finance'
finance_mapping = {'convenient': 1, 'inconv': 2}
dataset['finance'] = dataset['finance'].map(finance_mapping)

# Mapping for 'social'
social_mapping = {'nonprob': 1, 'slightly_prob': 2, 'problematic': 3}
dataset['social'] = dataset['social'].map(social_mapping)

# Mapping for 'health'
health_mapping = {'recommended': 1, 'priority': 2, 'not_recom': 3}
dataset['health'] = dataset['health'].map(health_mapping)
```
