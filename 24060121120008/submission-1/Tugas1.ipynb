## **1. Data yang digunakan**

```python
# Variabel URL menyimpan URL dari dataset yang diambil dari UCI Machine Learning Repository
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/liver-disorders/bupa.data"

# Variabel names berisi daftar nama kolom/atribut/feature yang akan digunakan untuk DataFrame yang akan dibaca.
names = ['mcv', 'alkphos', 'sgpt', 'sgot', 'gammagt', 'drinks', 'selector']
# KETERANGAN SETIAP FITUR:

# --- mvc ---
# mean corpuscular volume (mcv) adalah ukuran rata-rata volume sel darah merah.
# MCV dapat menunjukkan kondisi anemia. MCV yang rendah dapat mengindikasikan anemia defisiensi besi.
# Sedangkan MCV yang tinggi dapat mengindikasikan anemia defisiensi vitamin B12 atau folat.

# --- alkphos ---
# alkaline phosphotase (alkphos) adalah enzim yang ditemukan di hati, tulang, ginjal, dan usus.
# Alkphos berperan dalam memecah fosfat dari molekul lain.
# Alkphos yang tinggi dapat menunjukkan adanya kerusakan hati, penyumbatan saluran empedu, atau gangguan tulang.
# Alkphos yang rendah dapat menunjukkan adanya kekurangan seng, hipotiroidisme, atau malnutrisi.

# --- sgpt ---
# alanine aminotransferase (sgpt) adalah enzim yang ditemukan terutama di hati.
# SGPT berperan dalam mengubah asam amino alanin menjadi energi.
# SGPT yang tinggi dapat menunjukkan adanya kerusakan hati akibat hepatitis, sirosis, alkohol, obat-obatan, atau racun.
# SGPT yang rendah dapat menunjukkan adanya kekurangan vitamin B6, penyakit ginjal kronis, atau kehamilan.

# --- sgot ---
# aspartate aminotransferasee (sgot) adalah enzim yang ditemukan terutama di hati dan jantung.
# SGOT berperan dalam mengubah asam amino aspartat menjadi energi.
# SGOT yang tinggi dapat menunjukkan adanya kerusakan hati atau jantung akibat hepatitis, sirosis, alkohol, obat-obatan, racun, serangan jantung, atau gagal jantung.
# SGOT yang rendah dapat menunjukkan adanya kekurangan vitamin B6 atau penyakit ginjal kronis.

# --- gammagt ---
# gamma-glutamyl transpeptidase (gammagt) adalah enzim yang ditemukan di membran sel hati, ginjal, pankreas, dan jantung.
# Gammagt berperan sebagai molekul pembawa yang mengantarkan molekul lain ke seluruh tubuh.
# Gammagt juga berperan dalam membantu hati memetabolisme obat-obatan dan racun.
# Gammagt yang tinggi dapat menunjukkan adanya kerusakan hati, penyumbatan saluran empedu, radang pankreas, atau kanker kantong empedu.
# Gammagt yang rendah dapat menunjukkan adanya kekurangan seng, hipotiroidisme, atau malnutrisi.

# --- drinks ---
# number of half-pint equivalents of alcoholic beverages drunk per day (drinks) adalah jumlah minuman beralkohol yang dikonsumsi per hari dalam satuan setengah pint.
# Setengah pint setara dengan 0,3 liter atau 10 ons cair.
# Minuman beralkohol dapat mempengaruhi kesehatan hati, jantung, ginjal, otak, dan sistem saraf6

# --- selector ---
# field created by the BUPA researchers to split the data into train/test sets (selector) adalah variabel yang dibuat oleh peneliti BUPA untuk membagi data menjadi dua kelompok, yaitu data latih dan data uji.
# Data latih digunakan untuk melatih model klasifikasi atau regresi,
# sedangkan data uji digunakan untuk menguji kinerja model tersebut.
# Variabel ini tidak dapat digunakan sebagai variabel dependen atau target, karena tidak merepresentasikan adanya penyakit hati atau tidak.
# Variabel ini sering disalahartikan  oleh beberapa peneliti di masa lalu, sehingga menghasilkan hasil yang tidak valid dan berbahaya.

# Membaca dataset dari file CSV yang berada di URL dan menyimpannya dalam bentuk DataFrame.
# Kegunaannya adalah untuk memungkinkan analisis, pemrosesan, dan visualisasi data lebih lanjut
dataset = pd.read_csv(url, names=names)
```

**ii. Berikut merupakan bagian-bagian dari box and whisker plotâ–¶:**
1. Minimum: Nilai minimum dalam kumpulan data yang diberikan. Minimum adalah Q1 - 1.5 * IQR dan ditunjukkan oleh garis (whisker) sebelah bawah dari median (setelah kurtil ketiga).

2. Kuartil Pertama (Q1): Kuartil pertama adalah median setengah bawah dari kumpulan data.

3. Median: Median adalah nilai tengah dari kumpulan data, yang membagi kumpulan data yang diberikan menjadi dua bagian yang sama. Median dianggap sebagai kuartil kedua. Median ditunjukkan oleh garis yang berada di tengah kotak(box).

4. Kuartil Ketiga (Q3): Kuartil ketiga adalah median setengah atas dari data.

5. Maksimum: Nilai maksimum dalam kumpulan data yang diberikan. Maksimum adalah Q3 - 1.5 * IQR dan ditunjukkan oleh garis (whisker) sebelah atas dari median (sebelum kuartil pertama).

6. Rentang Antarkuartil (IQR): Perbedaan antara kuartil ketiga dan kuartil pertama dikenal sebagai rentang antarkuartil, yaitu IQR = Q3-Q1. IQR ditunjukkan oleh kotak(box).

7. Outlier: Data yang jatuh di sisi kiri atau kanan yang jauh dari data yang diurutkan diuji untuk menjadi outlier. Umumnya, outlier jatuh lebih dari jarak yang ditentukan dari kuartil pertama dan kuartil ketiga. Outlier ditunjukkan oleh lingkaran kecil yang berada diuar box and whisker. Outlier lebih besar dari Q3 + (1,5 . IQR) atau kurang dari Q1 - (1,5 . IQR).
