Digunakan untuk meng-*import* modul "matplotlib.pyplot" dan "seaborn" yang digunakan untuk membuat visualisasi data dalam bentuk grafik dan plot di Python.

```python
numeric_features = dataset.drop(['Class'], axis=1)
```

Digunakan untuk menghasilkan variabel bernama "numeric_features" yang berisi data numerik dengan menghapus kolom "Class" dari data asal.

```python
for column in numeric_features.columns:
    plt.figure(figsize=(8, 6))
    sns.histplot(numeric_features[column], kde=True)
    plt.title(f'Univariate Plot for {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

"for column ..." digunakan untuk melakukan iterasi melalui setiap kolom data dalam variabel "numeric_features".

"plt.figure(figsize=(8, 6))" digunakan untuk membuat plot dengan ukuran 8x6.

"sns.histplot ..." digunakan untuk menggambar histogram data dari kolom saat ini (saat iterasi berjalan) dengan garis *Kernel Density Estimation*.

"plt.title ..." digunakan untuk menambahkan judul plot yang sesuai dengan nama kolom saat ini (saat iterasi berjalan).

"plt.xlabel(column)" digunakan untuk menambahkan label sumbu x dengan nama kolom saat ini (saat iterasi berjalan).

"plt.ylabel('Frequency')" digunakan untuk menambahkan label "Frekuensi" pada sumbu y.

"plt.show()" digunakan untuk menampilkan plot.

```python
for column in numeric_features.columns:
    plt.figure(figsize=(8, 6))
    plt.hist(numeric_features[column], bins=20, edgecolor='k')
    plt.title(f'Histogram for {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numeric_columns = dataset.columns[1:]
```

Digunakan untuk menghasilkan variabel bernama "numeric_columns" yang berisi daftar nama kolom numerik dalam *dataset*, dimulai dari indeks kolom kedua (indeks 1) hingga kolom terakhir.

```python
for i in range(len(numeric_columns)):
    for j in range(i + 1, len(numeric_columns)):
        column1 = numeric_columns[i]
        column2 = numeric_columns[j]

        plt.figure(figsize=(6, 4))
        plt.scatter(dataset[column1], dataset[column2])
        plt.title(f'Scatter Plot: {column1} vs {column2}')
        plt.xlabel(column1)
        plt.ylabel(column2)
        plt.show()
```

Terdapat *loop* bersarang, di mana *loop* luar (dengan variabel i) berjalan melalui indeks kolom pertama dalam daftar "numeric_columns", dan *loop* dalam (dengan variabel j) berjalan melalui indeks kolom yang lebih tinggi dari i.

Variabel "column1" dan "column2" digunakan untuk menyimpan nama kolom yang akan digunakan dalam *scatter plot*.

"plt.figure(figsize=(6, 4))" digunakan untuk membuat plot dengan ukuran 6x4.

"plt.scatter(dataset[column1], dataset[column2])" digunakan untuk menggambar *scatter plot* (grafik titik-titik) antara dua kolom dalam "numeric_columns" yang sedang dianalisis.

"plt.title ..." digunakan untuk menambahkan judul ke plot yang mencerminkan nama dua kolom yang sedang dianalisis.

"plt.xlabel(column1)" digunakan untuk menambahkan label sumbu x dengan nama kolom yang sesuai (dengan iterasi).

"plt.ylabel(column1)" digunakan untuk menambahkan label sumbu y dengan nama kolom yang sesuai (dengan iterasi).

"plt.show()" digunakan untuk menampilkan *scatter plot* yang sedang dianalisis.
